```python
import numpy as np
import pandas as pd
import re
import ast
```

```python
main_df = pd.read_excel('holiday_packages_cleaned.xlsx')
main_df = main_df.drop('Unnamed: 0',axis=1)
```

```python
main_df
```

```python
main_df['cities_covered_with_duration'] = main_df['cities_covered_with_duration'].apply(lambda x: ast.literal_eval(x))
main_df['other_cities_covered'] = main_df['other_cities_covered'].fillna('[]').apply(lambda x: ast.literal_eval(x))
main_df['facilities'] = main_df['facilities'].apply(lambda x: ast.literal_eval(x))
main_df['sights_included'] = main_df['sights_included'].apply(lambda x: ast.literal_eval(x))
```

```python
def visit_to_remover(listt):
    L = []
    for i in listt:
        L.append(i.replace("Visit to",""))
    return L


main_df['sights_included'] = main_df['sights_included'].apply(visit_to_remover)
```

```python
df = main_df.copy()
```

```python
df
```

```python
def N_remover(cities):
    l = []
    for i in cities:
        i = re.sub(r'\d+N', '', i).strip()
        l.append(i)
    return l
```

```python
df['cities_covered_with_duration'] = df['cities_covered_with_duration'].apply(N_remover)
```

```python
df['other_cities_covered'] = df['other_cities_covered'].apply(N_remover)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df['cities_covered_with_duration'] = df['cities_covered_with_duration'].apply(collapse)
df['other_cities_covered'] = df['other_cities_covered'].apply(collapse)
df['facilities'] = df['facilities'].apply(collapse)
df['sights_included'] = df['sights_included'].apply(collapse)
```

```python
df['location'] = df['location'].apply(lambda x: [x])
```

```python
df['tags'] = df['location']+df['cities_covered_with_duration']+df['other_cities_covered']+df['facilities']+df['sights_included']
```

```python
final_df = df[['package_name','tags']]
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x: " ".join(x))
```

```python
final_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
vector = cv.fit_transform(final_df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(place):
    index = final_df[final_df['package_name'] == place].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(final_df.iloc[i[0]].package_name)
        
    
```

```python
recommend('Goa Delights - Free Catamaran Cruise')
```
